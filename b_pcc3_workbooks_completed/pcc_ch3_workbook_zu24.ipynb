## **WHAT IS A LIST**

```python
# nfl_teams = ['dolphins', 'jaguars', 'bucs', 'falcons', 'panthers']
# print(nfl_teams)

bicycles = ['trek', 'cannondale', 'redline', 'specialized']
# print(bicycles)

print(bicycles[0])
print(bicycles[0].title())
print(bicycles[0].upper())
```

### Print an element from a list

```python
bicycles = ['trek', 'cannondale', 'redline', 'specialized']

print(bicycles[0])
```

### Using Methods

```python
bicycles = ['trek', 'cannondale', 'redline', 'specialized']

print(bicycles[0])
print(bicycles[0].title())
print(bicycles[0].upper())

print() # empty line

# Print more than one item from a list
print(bicycles[1])
print(bicycles[3])

# prints the last index of a list
print(bicycles[-1])
```

### Using in Variables

```python
bicycles = ['trek', 'cannondale', 'redline', 'specialized']

message = f"My first bicycle was a {bicycles[0].title()}."
print(message)

msg1 = f"My first bike was a {bicycles[0].title()}, but now I have a {bicycles[2].title()}."
print(msg1)
```

## **MODIFYING, ADDING, REMOVING ELEMENTS**

### Modyfing Elements

```python
motorcycles = ['honda', 'yamaha', 'suzuki']
print(motorcycles)

motorcycles[0] = 'ducati'
print(motorcycles)

motorcycles[-1] = 'harley'
print(motorcycles)
```

### Adding Elements

```python
motorcycles = ['honda', 'yamaha', 'suzuki']

motorcycles.append('ducati')
print(motorcycles)
```

### Staring with an Empty List

```python
motorcycles = []

motorcycles.append('honda')
motorcycles.append('yamaha')
motorcycles.append('suzuki')

print(motorcycles)
```

### Inserting Elements

```python
motorcycles = ['honda', 'yamaha', 'suzuki']

motorcycles.insert(0, 'ducati')

print(motorcycles)
```

### Removing Elements (del)

```python
motorcycles = ['honda', 'yamaha', 'suzuki']

del motorcycles[2]

print(motorcycles)
```

### Removing Elements (pop)

```python
motorcycles = ['honda', 'yamaha', 'suzuki']

popped_motorcycle = motorcycles.pop()

print(motorcycles)
print(popped_motorcycle)


# starting over with the original list
motorcycles = ['honda', 'yamaha', 'suzuki']

last_owned = motorcycles.pop()
print(f"The last motorcycle I owned was a {last_owned.title()}.")

# starting over with the original list again
motorcycles = ['honda', 'yamaha', 'suzuki']

first_owned = motorcycles.pop(0)
print(f"The first motorcycle I owned was a {first_owned.title()}.")
```

### Removing Elements (remove)

```python
motorcycles = ['honda', 'yamaha', 'suzuki', 'ducati']

motorcycles.remove('ducati')

print(motorcycles)


# using the remove method and work with that value
motorcycles = ['honda', 'yamaha', 'suzuki', 'ducati']
print(motorcycles)

too_expensive = 'ducati'
motorcycles.remove(too_expensive)
print(motorcycles)
print(f"A {too_expensive.title()} is too expensive for me.")
```

## **ORGANIZING ELEMENTS**

### Sorting a list permanently

```python
cars = ['bmw', 'audi', 'toyota', 'subaru']
cars.sort()
print(cars)

# reverse order
cars.sort(reverse=True)
print(cars)

# back to alphabetical
cars.sort()
print(cars)
```

### Sorting a list temporarily

```python
cars = ['bmw', 'audi', 'toyota', 'subaru']

print("Here is the original list:")
print(cars)

print() # emptyline

print("Here is the sorted list:")
print(sorted(cars))
print("Here is the reverser sorted list:")
print(sorted(cars, reverse=True))

print() # emptyline

print("Here is the original list again:")
print(cars)
```

### Reverse Order (permanent)

```python
cars = ['bmw', 'audi', 'toyota', 'subaru']

# reverse order permamently
print(cars)
cars.reverse()
print(cars)
```

### Finding the Length


```python
cars = ['bmw', 'audi', 'toyota', 'subaru']

print(len(cars))
```
