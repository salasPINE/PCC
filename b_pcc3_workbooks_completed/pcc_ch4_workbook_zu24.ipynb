## LOOPING THROUGH AN ENTIRE LIST

```python
# magicians.py

magicians = ['alice', 'david', 'carolina']
for magician in magicians:
  print(magician)
print("We're done here!")

print()

cats = ['mittens', 'milo', 'garfield']
for cat in cats:
  print(cat)

print()

dogs = ['coco', 'charlie', 'enzo']
for dog in dogs:
  print(dog)
```

## AVOIDING INDENTATION ERRORS

### Forgetting to Indent.

```python
magicians = ['alice', 'david', 'carolina']
for magician in magicians:
print(magician)
# the print statement should be indented
```

### Forgetting to Indent Additional Lines

```python
magicians = ['alice', 'david', 'carolina', 'zoltar']
for magician in magicians:
  print(f"{magician.title()}, that was a great trick!")
print(f"I can't wait to see your next trick, {magician.title()}.")
# this code will work, but it is not what we wanted
# the second print statement should be indented
```

### Indenting Unnecessarily

```python
magicians = ['alice', 'david', 'carolina', 'zoltar']
for magician in magicians:
  print(f"{magician.title()}, that was a great trick!")
  print(f"I can't wait to see your next trick, {magician.title()}.")

  print("Thank you everyone, that was a great magic show!")
# this code will work, but it is not what we wanted
# the last print statement should NOT be indented
# it should be outside the loop
```

## MAKING NUMBERICAL LISTS

### Using the `range()` Function


```python
# first_numbers.py

for value in range(1, 6):
  print(value)

print()
for value in range(6):
  print(value)
```

### Using `range()` to Make a List of Numbers

```python
numbers = list(range(1, 6))
print(numbers)

even_numbers = list(range(2, 11, 2))
print(even_numbers)

odd_numbers = list(range(1, 10, 2))
print(odd_numbers)

print()

# square_numbers.py
squares = []
for value in range(1, 11):
  square = value ** 2
  squares.append(square)

print(squares)

# cube_numbers.py
cubes = []
for value in range(1, 11):
  cube = value ** 3
  cubes.append(cube)

print(cubes)
```

### List Comprehensions

```python
# allows you to create a list with computations in one line
squares = [value**2 for value in range(1, 11)]
print(squares)
```

### Simple Statistics with a List of Numbers

```python
digits = [1,2,3,4,5,6,7,8,9,0]
print(min(digits))
print(max(digits))
print(sum(digits))

digits2 = list(range(10))
print(min(digits2))
print(max(digits2))
print(sum(digits2))
```

## WORKING WITH PART OF A LIST

### Slicing a List

```python
# players.py

players = ['charles', 'martina', 'michael', 'lionel', 'andre']
print(players[0:3])
print(players[1:4])

# omit the first index in the slice
print(players[:4])
# omit the second index in the slice
print(players[2:])
# print the last three players
print(players[-3:])
```

### Looping Through a Slice

```python
players = ['charles', 'martina', 'michael', 'lionel', 'andre']

print("Here are the first three players on my team:")
for player in players[:3]:
  print(player.title())
```

### Copying a List

```python
# To copy a list, make a slice that includes the entire original list
# foods.py
my_foods = ['pizza', 'falafel', 'carrot cake']
friend_foods = my_foods[:]

my_foods.append('canoli')
friend_foods.append('ice cream')

print("My favorite foods are:")
print(my_foods)
print()
print("My friends favorite foods are:")
print(friend_foods)
print()

# this does not work!!
friend_foods = my_foods
```

## TUPLES

### Defining a Tuple

```python
# a tuple uses regular parenthese instead of square brackets

dimensions = (200, 50)
print(dimensions)
print(dimensions[0])
print(dimensions[-1])

# try to change an item in a tuple - won't work
dimensions[0] = 250
```

### Looping Through All Values in a Tuple

```python
dimensions = (200, 50)
for value in dimensions:
  print(value)
```

### Writing Over a Tuple

```python
dimensions = (200,50)
for value in dimensions:
  print(value)

dimensions = (400,100)
print("New dimensions!")
for value in dimensions:
  print(value)
```

## STYLING YOUR CODE

PEP 8 instructs programmers on how to style their code. Code will always be "read" more often than it is written. Always write code that is easy to read.

### Indentation
* Use four spaces per indentation level
* Do not use TABS

### Line Length
* Each line should only be 80 characters per line
* Each comment should only be 72 characters per line
* 99 character lines are also common

### Blank Lines
* Use blank lines to group parts of your program visually.
* Do not use multiple blank lines 3-4 between sections

### Comments
* Always use comments above your code to explain what is happening
