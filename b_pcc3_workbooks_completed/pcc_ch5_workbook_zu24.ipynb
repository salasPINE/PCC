```python
car = 'bmw'
car == 'bmw'

car = 'audi'
car == 'bmw'

car = 'Audi'
car == 'audi'

car = 'AUDI'
car.lower() == 'audi'

print(car)
```

```python
age = 18
age == 18

answer = 17
if answer != 42:
    print("That is not the correct answer. Try again!")

age = 19
age < 21

age <= 21

age >= 21

age > 21
```

```python
age_0 = 22
age_1 = 18

age_0 >= 21 and age_1 >= 21

age_0 = 22
age_1 = 22
age_0 >= 21 and age_1 >= 21

(age_0 >= 21) and (age_1 >= 21)
```

```python
age_0 = 22
age_1 = 18

(age_0 >= 21) or (age_1 >= 21)

age_0 = 18
(age_0 >= 21) or (age_1 >= 21)
```

## **`if` Statements**

#### **Simple** `if` **Statements**

```python
# the simplest IF statement
# if conditional_test:
#    do something

# voting.py
age = 19
if age >= 18:
    print("You are old enough to vote!")
    print("Have you registered to vote yet?")
```

#### `if-else` **Statements**

```python
age = 12
if age >= 18:
    print("You are old enough to vote!")
    print("Have you registered to vote yet?")
else:
    print("Sorry, you are too young to vote.")
    print("Please register to vote when you are 18.")
```

#### **The** `if-elif-else` **Chain**

```python
# amusement_park.py

age = 2

if age < 4:
    print("Your admission cost is $0")
elif age < 18:
    print("Your admission cost is $25")
else:
    print("Your admission cost is $40")
```

#### **Using Multiple** `elif` **Blocks**

```python
# amusement_park.py
# with senior discount

age = 87

if age < 4:
    print("Your admission cost is $0")
elif age < 18:
    print("Your admission cost is $25")
elif age < 65:
    print("Your admission cost is $40")
else:
    print("Your admission cost is $20")
```

#### **Ommitting the** `else` **Block**

```python
# amusement_park.py
# you can omit the last else:
# and replace it with another elif

age = 87

if age < 4:
    print("Your admission cost is $0")
elif age < 18:
    print("Your admission cost is $25")
elif age < 65:
    print("Your admission cost is $40")
elif age >= 65:
    print("Your admission cost is $20")
```

#### **Testing Multiple Conditions**

```python
# toppings.py
# checking multiple conditions

requested_toppings = ['mushrooms', 'extra cheese']

if 'mushrooms' in requested_toppings:  # True
    print("Adding mushrooms!")
if 'pepperoni' in requested_toppings:  # False
    print("Adding pepperoni!")
if 'extra cheese' in requested_toppings:  # True
    print("Adding extra cheese!")

print()

print("Finished making your pizza!")
```

## **Using `if` Statements with Lists**

#### **Checking for Special Items**

```python
# toppings.py

requested_toppings = ['mushrooms', 'green peppers', 'extra cheese']

for topping in requested_toppings:
    print(f"Adding {topping}.")

print()
print("Finished making your pizza!")
print()

# what if the pizzeria runs out of green peppers?
# use an if statement inside the loop to check

for topping in requested_toppings:
    if topping == 'green peppers':
        print("Sorry, we are out of green peppers.")
    else:
        print(f"Adding {topping}")
print()
print("Finished making your pizza!")
```

#### **Checking That a List Is Not Empty**

```python
requested_toppings = []  # empty list

if requested_toppings:
    for topping in requested_toppings:
        print(f"Adding {topping}")
    print("Finished making your pizza!")
else:
    print("Are you sure you want a plain pizza?")
```

#### **Using Multiple Lists**

```python
# our pizza toppings
available_toppings = ['mushrooms', 'olives', 'green peppers',
                      'pepperoni', 'pineapple', 'extra cheese']

# the toppings being asked for
requested_toppings = ['mushrooms', 'french fries', 'extra cheese']

for topping in requested_toppings:
    if topping in available_toppings:
        print(f"Adding {topping}")
    else:
        print(f"Sorry, we don't have {topping}")

print()
print("Finished making your pizza!")
```
