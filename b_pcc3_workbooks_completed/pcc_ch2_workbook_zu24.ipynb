## **STRINGS**

```python
# Strings are in quotes. They are any text or numbers that you want to use.
print("This is a string")
print('This is also a string')
print('"This is quoted"')
print('I told my friend, "Python is my favorite language!"')
print("The language 'Python' is named after Monty Python, not the snake.")

# You can use both types of quotes if necessary.
# But always use the same ones at the beginning and end of the string
print("I Can't Do This Anymore!")
```

### Changing Case

```python
# try to use lowercase for strings
name = "ada lovelace"
print(name)

# use a method to change the case of the string in the variable
print(name.title())
print(name.upper())

name = "ADA LOVELACE"
print(name.lower())
```

### Using Variables in Strings

```python
# this code shows how to embed a variable in a string using f strings
first_name = "ada"
last_name = "lovelace"

# always start with the f and then put the variable in curly brackets
full_name = f"{first_name} {last_name}"

# you can also have a combination of variables, methods and typed in a string
print(f"Hello, {full_name.title()}!")

message = f"Hello, {full_name.title()}!"

print(message)
```

### Adding Whitespace

```python
# whitespace is non-printing characters, invisible, but they are there
# \t is a tab (4 spaces), \n is a new line, like the return key
print("Python")
print("\tPython")

print("Languages:\nPython\nJava\nJavascript")
```

### Stripping Whitespace

```python
# use a method in the variable to get rid of whitespace
# to the right, the left, or both sides
favorite_language = '     python     '
print(favorite_language)

print(favorite_language.rstrip())
print(favorite_language.lstrip())

print(favorite_language.strip())
```

### Removing Prefixes and Suffixes

```python
# use a method for removing specific text from the start or end of a string
# start = prefix, end = suffix
# you need to put what you want to remove in the parentheses using single quotes
url = "https://www.thepineschool.org"

print(url.removeprefix('https://'))
print(url.removeprefix('https://www.'))

filename = "python_notes.txt"
codefile = "python_code.py"

print(filename.removesuffix('.txt'))
print(codefile.removesuffix('.py'))
```

## **NUMBERS**

###Integers

```python
# integers are whole numbers
print(2 + 2)
print(3 - 2)
print(3 * 3)
print(3 / 2)
print(3 ** 2)
```

### Floats

```python
# floats = floating point numbers
# floats have a decimal point 1.5, 0.7, etc.
# division always uses floating point numbers
print(0.2 + 0.2)
print(2 * 0.1)
```

### Integers and Floats

```python
print(4 * 2)
print(4 / 2)
```

### Underscores in Numbers

```python
# you can use underscores in the place of commas to make big numbers easier to read
# they are ignored by Python
universe_age = 14_000_000_000
print(universe_age)
```

### Multiple Assignment

```python
# you can assign multiple variables at once on one line
x, y, z = 1, 2, 3

print(y)
print(x)
print(z)

print(f"{x},{y},{z}")
```

### Constants

```python
# constants are done in ALL CAPS. it is a visual indicator
MAX_CONNECTIONS = 5000
print(MAX_CONNECTIONS)

MAX_CONNECTIONS = 4000
print(MAX_CONNECTIONS)
```

### Difference Between Strings and Integers

```python
print(3)
print("3")
```
